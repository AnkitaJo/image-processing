```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import imresize
from skimage.exposure import rescale_intensity
%matplotlib inline 
from numpy import array, zeros
import scipy.misc
```

```python
image = cv2.imread("input/lenna.png",0)

img = imresize(image, (1024,1024))

size = img.shape

#level 2
size1 = (size[0]/2, size[1]/2)
r1=np.zeros((size1[0],size1[1]),dtype=np.integer)
m=0
for i in range(0,size[0],2):
    n=0
    for j in range (0,size[1],2):
        r1[m][n]= (img[i][j]/4) + (img[i+1][j]/4) + (img[i][j+1]/4) + (img[i+1][j+1]/4)
        n = n+1
    m = m+1
        
#level 3
size2 = (size1[0]/2, size1[1]/2)
r2=np.zeros((size2[0],size2[1]),dtype=np.integer)
m=0
for i in range(0,size1[0],2):
    n=0
    for j in range (0,size1[1],2):
        r2[m][n]= (r1[i][j]/4) + (r1[i+1][j]/4) + (r1[i][j+1]/4) + (r1[i+1][j+1]/4)
        n = n+1
    m = m+1
#from skimage.exposure import rescale_intensityfrom skimage.exposure import rescale_intensity
#level 4
size3 = (size2[0]/2, size2[1]/2)
r3=np.zeros((size3[0],size3[1]),dtype=np.integer)
m=0
for i in range(0,size2[0],2):
    n=0
    for j in range (0,size2[1],2):
        r3[m][n]= (r2[i][j]/4) + (r2[i+1][j]/4) + (r2[i][j+1]/4) + (r2[i+1][j+1]/4)
        n = n+1
    m = m+1

#level 5
size4 = (size3[0]/2, size3[1]/2)
r4=np.zeros((size4[0],size4[1]),dtype=np.integer)
m=0
for i in range(0,size3[0],2):
    n=0
    for j in range (0,size3[1],2):
        r4[m][n]= (r3[i][j]/4) + (r3[i+1][j]/4) + (r3[i][j+1]/4) + (r3[i+1][j+1]/4)
        n = n+1
    m = m+1
    
#level 6
size5 = (size4[0]/2, size4[1]/2)
r5=np.zeros((size5[0],size5[1]),dtype=np.integer)
m=0
for i in range(0,size4[0],2):
    n=0
    for j in range (0,size4[1],2):
        r5[m][n]= (r4[i][j]/4) + (r4[i+1][j]/4) + (r4[i][j+1]/4) + (r4[i+1][j+1]/4)
        n = n+1
    m = m+1
    

scipy.misc.imsave('output/lenna1.png', r1)
scipy.misc.imsave('output/lenna2.png', r2)
scipy.misc.imsave('output/lenna3.png', r3)
scipy.misc.imsave('output/lenna4.png', r4)
scipy.misc.imsave('output/lenna5.png', r5)


```

```python
#Zero order hold scheme: For 32x32
mask1 = np.array((
    [1, 1],
    [1, 1]), dtype="int")


#starting from level 5
s = r5.shape
up1 = zeros((s[0]*2+1,s[1]*2+1))
s1 = up1.shape
for i in range(s[0]):
    for j in range(s[1]):
        up1[i*2+1][j*2+1] = r5[i][j]

dst = cv2.filter2D(up1,-1,mask1)

s = dst.shape
up1 = zeros((s[0]*2+1,s[1]*2+1))
s1 = up1.shape
for i in range(s[0]):
    for j in range(s[1]):
        up1[i*2+1][j*2+1] = dst[i][j]
        
dst2 = cv2.filter2D(up1,-1,mask1)


s = dst2.shape
up1 = zeros((s[0]*2+1,s[1]*2+1))
s1 = up1.shape
for i in range(s[0]):
    for j in range(s[1]):
        up1[i*2+1][j*2+1] = dst2[i][j]
        
dst3 = cv2.filter2D(up1,-1,mask1)

s = dst3.shape
up1 = zeros((s[0]*2+1,s[1]*2+1))
s1 = up1.shape
for i in range(s[0]):
    for j in range(s[1]):
        up1[i*2+1][j*2+1] = dst3[i][j]
        
dst4 = cv2.filter2D(up1,-1, mask1)

s = dst4.shape
up1 = zeros((s[0]*2+1,s[1]*2+1))
s1 = up1.shape
for i in range(s[0]):
    for j in range(s[1]):
        up1[i*2+1][j*2+1] = dst4[i][j]
        
dst5 = cv2.filter2D(up1,-1, mask1)

plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(r5, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(dst5, cmap='gray')
plt.show()


#scipy.misc.imsave('output/lenna11.png', dst4)
```
